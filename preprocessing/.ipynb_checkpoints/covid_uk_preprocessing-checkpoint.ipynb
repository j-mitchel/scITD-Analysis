```r
library(Seurat)
library(SeuratDisk)
```

```r
# data downloaded from: https://www.covid19cellatlas.org/index.patient.html
# Titled “COVID-19 PBMC Ncl-Cambridge-UCL”
```

```r
# convert h5ad to h5seurat
Convert("/home/jmitchel/data/covid_data_uk/haniffa21.processed.h5ad", dest = "h5seurat", overwrite = FALSE)
```

```r
# read the h5seurat file
pbmc <- LoadH5Seurat("/home/jmitchel/data/covid_data_uk/haniffa21.processed.h5seurat")
```

```r
# qc analyses
pbmc[['nCount_RNA']] <- colSums(pbmc@assays$raw@counts)
# pbmc[["percent.mt"]] <- PercentageFeatureSet(pbmc, pattern = "^MT-")
pbmc[['percent.mt']] <- PercentageFeatureSet(pbmc, pattern = "^MT-")
```

```r
# Visualize QC metrics as a violin plot
VlnPlot(pbmc, features = c("nCount_RNA", "percent.mt"), ncol = 2)
```

```r
# look at metadata columns
colnames(pbmc@meta.data)
```

```r
# see how many samples and donors there are
print(length(levels(pbmc@meta.data$patient_id)))
print(length(levels(pbmc@meta.data$sample_id)))
```

```r
# determine how many samples per donor there are
donors <- levels(pbmc@meta.data$patient_id)
for (d in donors) {
    cells_keep <- rownames(pbmc@meta.data)[pbmc@meta.data$patient_id==d]
    pbmc_sub <- subset(pbmc,cells=cells_keep)
    print(unique(pbmc_sub@meta.data$sample_id))
    if (length(unique(pbmc_sub@meta.data$sample_id))>1) {
        print(unique(pbmc_sub@meta.data$Collection_Day))
        print(unique(pbmc_sub@meta.data$Days_from_onset))
        print(unique(pbmc_sub@meta.data$Days_from_onset))
    }
}
# seems like there are 13 donors that have two samples (all from Cambridge cohort)
# they had 1 sample taken D0 and generally the second D28, except for the one healthy donor
# which had labels only of D0 since there is no notion of progression with this individual
```

```r
unique(pbmc@meta.data[pbmc@meta.data$patient_id=='CV0284',c('sample_id','patient_id','Site','Days_from_onset','Collection_Day')])
```

```r
unique(pbmc@meta.data[pbmc@meta.data$Collection_Day=='D28',c('sample_id','patient_id','Site','Days_from_onset','Collection_Day')])
```

```r
## I should keep only the D0 samples for all donors
cells_keep <- rownames(pbmc@meta.data)[pbmc@meta.data$Collection_Day=='D0']
print(length(cells_keep))
```

```r
pbmc_d0 <- subset(pbmc,cells=cells_keep)
```

```r
# for the one healthy donor with two D0 samples, just arbitrarily keep one of them
# will arbitrarily remove this one BGCV02_CV0902
cells_keep <- rownames(pbmc_d0@meta.data)[pbmc_d0@meta.data$sample_id!='BGCV02_CV0902']
print(length(cells_keep))
```

```r
pbmc_d0_2 <- subset(pbmc_d0,cells=cells_keep)
```

```r
# remove the patients with other immune diseases
levels(pbmc_d0_2@meta.data$Status)

cells_keep <- rownames(pbmc_d0_2@meta.data)[pbmc_d0_2@meta.data$Status!='Non_covid']
print(length(cells_keep))
pbmc_d0_3 <- subset(pbmc_d0_2,cells=cells_keep)
print(length(levels(pbmc_d0_3@meta.data$patient_id)))
```

```r
print(length(unique(pbmc_d0_3@meta.data$patient_id))) #when doing levels above it still includes those that don't have any cells left
```
