```r
library(splatter)
library("scater")
library(devtools)
```

```r
setwd('/home/jmitchel/scITD')
load_all()
```

```r
vcf <- mockVCF(n.samples = 20)
gff <- mockGFF()
```

```r
dim(vcf)
```

```r
# set sim params
params.cond <- newSplatPopParams(similarity.scale = 5,
                                 condition.prob = c(0.5, 0.5),
                                 cde.facLoc = 0.5, 
                                 cde.facScale = 0.5,
                                 de.prob = 0.5,
                                 de.facLoc = 0.5, 
                                 de.facScale = 0.5,
                                 group.prob = c(0.5, 0.5),
                                 batchCells = c(50),
                                 batch.size = 20, nGenes=1000)

params.cond

```

```r
# generate sim
sim.pop.cond <- splatPopSimulate(vcf = vcf, gff = gff,
                                 params = params.cond, 
                                 sparsify = FALSE)
```

```r
# visualize sim data
sim.pop.cond <- logNormCounts(sim.pop.cond)
sim.pop.cond <- runPCA(sim.pop.cond, ncomponents = 10)
plotPCA(sim.pop.cond, colour_by = "Sample", shape_by = "Condition", point_size=3)

```

```r
# prep data for use with scITD
simcounts <- counts(sim.pop.cond)
simcounts <- methods::as(as.matrix(simcounts),'sparseMatrix')
simmeta <- as.data.frame(colData(sim.pop.cond)@listData)
colnames(simmeta)[4] <- 'donors'
colnames(simmeta)[5] <- 'ctypes'
```

```r
param_list <- initialize_params(ctypes_use = c("Group1", "Group2"),
                                ncores = 30, rand_seed = 10)

sim_container <- make_new_container(count_data=simcounts, meta_data=simmeta, params=param_list)

sim_container <- form_tensor(sim_container, donor_min_cells=5,
                              norm_method='trim', scale_factor=10000,
                              vargenes_method='norm_var_pvals', vargenes_thresh=1,
                              scale_var = TRUE, var_scale_power = .5)

# sim_container <- run_tucker_ica(sim_container, ranks=c(2,4),
#                                  tucker_type = 'regular', rotation_type = 'ica_dsc')
sim_container <- run_tucker_ica(sim_container, ranks=c(2,4),
                                 tucker_type = 'regular', rotation_type = 'hybrid')

```

```r
sim_container <- plot_donor_matrix(sim_container, 
                                    show_donor_ids = TRUE)

sim_container$plots$donor_matrix
```
