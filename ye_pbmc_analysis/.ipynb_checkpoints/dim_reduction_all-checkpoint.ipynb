```r
library(conos)
library(Seurat)
library(SeuratDisk)
library(sccore)
library(pagoda2)
```

```r
pbmc <- readRDS('/home/jmitchel/data/lupus_data/lupus_subsetted_seurat_v3.rds')
```

```r
all_donors <- unique(pbmc@meta.data$ind_cov_batch_cov)
```

```r
length(all_donors)
```

```r
# make the conos object
all_d_data <- list()
for (d in all_donors) {
    cells_keep <- rownames(pbmc@meta.data)[pbmc@meta.data$ind_cov_batch_cov==d]
    pbmc_sub <- subset(pbmc,cells=cells_keep)
    all_d_data[[d]] <- pbmc_sub@assays$RNA@counts
}

# turn the list of matrices to list of pagoda2 objects
panel.preprocessed <- lapply(all_d_data, pagoda2::basicP2proc, n.cores=5,
                             min.cells.per.gene=0, n.odgenes=2e3,
                             get.largevis=FALSE, make.geneknn=FALSE)
```

```r
saveRDS(panel.preprocessed,file='/home/jmitchel/data/lupus_data/lupus_pagoda_panel.rds')
```

```r
length(panel.preprocessed)
```

```r
con <- conos::Conos$new(panel.preprocessed, n.cores=5)

# build graph
con$buildGraph()

# make umap embedding
con$embedGraph(method="UMAP", min.dist=0.01, spread=15, min.prob.lower=1e-3)

```
